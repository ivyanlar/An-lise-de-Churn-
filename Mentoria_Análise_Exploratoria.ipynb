De acordo com a literatura, podemos utilizar algumas técnicas para corrigir os missings: 
  - Excluir valores 
  - Utilizar a mediana 
  - Utilizar o valor mais frequente 

  No entando, a remoção de dados não se torna adequada pois a variável **reamining_contract** exixtem 21.572 observações. Se removidas, enquanto que as outras variáveis, existem 381 observações ausentes.

```python
#Para corrigir os dados ausentes em reamining_contract, irei utilizar a mediana, porque se trata de uma varável numérica
```

## Medir a associação entre variáveis categóricas -- **Teste de Independencia**

H0: As variáveis de classificação são independentes; 

H1: As variáveis de classificação não são independentes;

```python
!pip install scipy
# Legenda:
# is_tv_subscriber - o cliente tem uma tv por assinatura 
# is_movie_package_ - assinatura de pacote de cinema
# download_over_limit- se o limite dos downloads foi atingindo
# churn - se houve o cancelamento ou não
```

```python
## Primeiro teste:
# as variáveis "is_tv_subscriber" e "is_movie_package" são independentes
import pandas as pd 
ctab1 = pd.crosstab(tabela_total['is_tv_subscriber'],tabela_total['is_movie_package_subscriber'])
import scipy.stats as stats
stats.chi2_contingency(ctab1)
chi, p, dof, exp = stats.chi2_contingency(ctab1) 
print(chi, p)

# O valor p é inferior ao nível de siginificância, logo a decisão é de rejeitar a HO.
## Portanto, as variáveis "is_tv_subscriber" e "is_movie_package" são depentes.
```

```python
## segundo teste:
# as variáveis "is_tv_subscriber" e "download_over_limit" são independentes
ctab2 = pd.crosstab(tabela_total['is_tv_subscriber'],tabela_total['download_over_limit'])
import scipy.stats as stats
stats.chi2_contingency(ctab2)
chi, p, dof, exp = stats.chi2_contingency(ctab2) 
print(chi, p)
# O valor p é inferior ao nível de siginificância, logo a decisão é de rejeitar a HO.
## Portanto, as variáveis "is_tv_subscriber" e "is_movie_package" são depentes.
```
